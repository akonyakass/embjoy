Data Processing

```python
df_long = pd.melt(df, id_vars=['ISO3', 'Country', 'Continent', 'Hemisphere', 'Human Development Groups', 'UNDP Developeing Regions', 'HDI Rank (2021)'],
                  var_name='Maternal Mortality Ratio Year', value_name='Maternal Mortality Ratio')
df_long
```

```python
df_long['Years'] = df_long['Maternal Mortality Ratio Year'].str.extract('(\d{4})').astype(float)
df_long.fillna(0, inplace=True)
df_long
```

Mortality Analysis across continents over years

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x='Years', y='Maternal Mortality Ratio', hue='Continent')
plt.title('Maternal Mortality Ratio Trends (1990-2021) by Continent')
plt.xlabel('Year')
plt.ylabel('Maternal Mortality Ratio')
plt.show()
```

Mortality Analysis - Continents

```python
# Group by Continent and calculate the mean maternal mortality ratio for each continent
continent_mean = df_long.groupby('Continent')['Maternal Mortality Ratio'].mean().reset_index()

# Bar plot for continent-wise average maternal mortality ratio
plt.figure(figsize=(12, 6))
sns.barplot(data=continent_mean, x='Continent', y='Maternal Mortality Ratio')
plt.title('Average Maternal Mortality Ratio by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Maternal Mortality Ratio')
plt.show()
```

Mortality Analysis - Countries

```python
country_mean = df_long.groupby('Country')['Maternal Mortality Ratio'].mean().reset_index()

# Bar plot for country-wise average maternal mortality ratio (top 10 countries for example)
top_countries = country_mean.sort_values(by='Maternal Mortality Ratio', ascending=False).head(50)
plt.figure(figsize=(12, 12))
sns.barplot(data=top_countries, x='Maternal Mortality Ratio', y='Country')
plt.title('Top 50 Countries with Highest Maternal Mortality Ratio')
plt.xlabel('Average Maternal Mortality Ratio')
plt.ylabel('Country')
plt.show()
```

HDI Rank of Countries Over Years

```python
# Calculate the mean HDI Rank (2021) for each country
country_mean_hdi = df_long.groupby('Country')['HDI Rank (2021)'].mean().reset_index()

# Select the top 10 or 15 countries based on mean HDI Rank (2021)
top_countries = country_mean_hdi.nlargest(15, 'HDI Rank (2021)')

# Filter the DataFrame to include only the top countries
df_top_countries = df_long[df_long['Country'].isin(top_countries['Country'])]

# Bar plot to visualize the trend of HDI Rank (2021) for the top countries over the years
plt.figure(figsize=(12, 8))
sns.barplot(data=df_top_countries, x='Country', y='HDI Rank (2021)', hue='Years', palette='viridis')
plt.title('Trend of HDI Rank (2021) for Top Countries Over the Years')
plt.xlabel('Country')
plt.ylabel('HDI Rank (2021)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Years')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

Maternal Mortality Ratio VS Human Development Groups

```python
# Box plot to compare Maternal Mortality Ratio across Human Development Groups
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_long, x='Human Development Groups', y='Maternal Mortality Ratio', palette='Set3')
plt.title('Maternal Mortality Ratio Across Human Development Groups')
plt.xlabel('Human Development Groups')
plt.ylabel('Maternal Mortality Ratio (deaths per 100,000 live births)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Box plot to compare Maternal Mortality Ratio across UNDP Developing Regions
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_long, x='UNDP Developeing Regions', y='Maternal Mortality Ratio', palette='Set2')
plt.title('Maternal Mortality Ratio Across UNDP Developing Regions')
plt.xlabel('UNDP Developeing Regions')
plt.ylabel('Maternal Mortality Ratio (deaths per 100,000 live births)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
